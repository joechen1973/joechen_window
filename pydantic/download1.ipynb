```python

#可以從json檔案中抓出自己想要的資料
#pydantic是確認資料的安全性,以避免缺資料,或....

import requests
from requests import Response

aqi_url = 'https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate desc&format=JSON'

# res:requests.Response = requests.request("GET",aqi_url)
res:requests = requests.request("GET",aqi_url)
if res.ok:
    print("下載成功")
else:
    print("下載失敗")
```

```python
# print(res.text)
type(res.text)
```

```python
json_data:dict = res.json()
sites:list = json_data["records"]
for site in sites:
    print(site)
```
